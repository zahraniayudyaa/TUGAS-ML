# **1. Apa itu Machine Learning?**
---
Machine Learning (ML) atau Pembelajaran Mesin adalah cabang dari Kecerdasan Buatan (AI) yang memungkinkan sistem komputer untuk belajar dari data tanpa diprogram secara eksplisit. Intinya, alih-alih memberikan instruksi langkah demi langkah, kita memberi mesin data dan algoritma yang memungkinkannya meningkatkan kinerjanya melalui pengalaman.
Konsep Utama :

* Belajar dari Data, ML menggunakan data (angka, teks, gambar, dll.) untuk mengenali pola, membuat prediksi, atau mengambil keputusan.
* Meningkatkan dengan Pengalaman,
Semakin banyak data yang dilatih, model ML biasanya semakin akurat.

# **2. Mengapa Menggunakan Machine Learning?**
---
Menggunakan machine learning (ML) memberikan keunggulan saat menyelesaikan masalah yang sulit/tidak praktis diprogram secara manual. Berikut alasan utamanya:

* Untuk Masalah yang Terlalu Kompleks bagi Pemrograman Tradisional
* Ketika Data Terlalu Besar atau Dinamis untuk Dikelola Manual
* Untuk Personalisasi dalam Skala Besar
* Mengotomatisasi Tugas yang Membosankan dan Berulang
* Membuat Prediksi Masa Depan Berbasis Data Historis
* Untuk Masalah yang Melibatkan Persepsi Manusia (Computer Vision, NLP)
* Ketika Solusi Perlu Terus Beradaptasi dan Meningkatkan Diri

# **4. Jenis Sistem Machine Learning**
---
###**A. Berdasarkan Supervisi (Supervision)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1.&nbsp;&nbsp;Supervised Learning (Data memiliki label/target)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Klasifikasi (spam/ham), Regresi (prediksi harga)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Algoritma:** Linear Regression, Logistic Regression, SVM, Decision Trees

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2.&nbsp;&nbsp;Unsupervised Learning (Data tanpa label)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Clustering, Dimensionality reduction, Anomaly detection  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Algoritma:** K-Means, PCA, t-SNE

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;3.&nbsp;&nbsp;Semi-supervised Learning (Kombinasi data berlabel dan tidak berlabel)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Google Photos yang mengenali wajah

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;4.&nbsp;&nbsp;Reinforcement Learning (Agent belajar dari interaksi dengan environment)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** AlphaGo, robot belajar berjalan
<br><br>

###**B. Berdasarkan Kemampuan Incremental Learning**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1.&nbsp;&nbsp;Batch Learning (Belajar dari seluruh data sekaligus dan training dilakukan offline)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Sistem rekomendasi yang di-update mingguan

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2.&nbsp;&nbsp;Online Learning (Belajar secara incremental dari data stream dan dapat beradaptasi secara real-time)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Prediksi harga saham, filter spam adaptif
<br><br>
###**C. Berdasarkan Metode Generalisasi**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1.&nbsp;&nbsp;Instance-based Learning
(Menghafal contoh training dengan generalisasi berdasarkan kemiripan)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** k-Nearest Neighbors (k-NN)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2.&nbsp;&nbsp;Model-based Learning (Membangun model matematis dari data serta mencari parameter optimal)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Linear Regression, Neural Networks

# **5. Tantangan Utama Machine Learning**
---
###**A. Data yang Tidak Cukup (Insufficient Data)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ML membutuhkan data dalam jumlah besar untuk belajar efektif dan adanya "Unreasonable Effectiveness of Data", dimana terkadang data lebih penting daripada algoritma kompleks.  
**Aturan praktis:** Ribuan contoh untuk masalah sederhana, jutaan untuk masalah kompleks

###**B. Data Training yang Tidak Representatif**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Training data harus mewakili kasus baru yang akan dihadapi.  
**Sampling bias:** Contoh terkenal, yaitu polling pemilihan US 1936 yang salah prediksi karena hanya survey orang kaya  
**Nonresponse bias:** Hanya orang tertentu yang merespon survey

###**C. Data Berkualitas Buruk (Poor-Quality Data)**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Data dengan noise, outlier, atau missing values.  
**Teknik:** imputasi missing values, outlier removal, data normalization

###**D. Fitur yang Tidak Relevan (Irrelevant Features)**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;"Garbage in, garbage out"  

Feature engineering kritis untuk sukses:  
* Feature selection: Memilih fitur paling informatif  
* Feature extraction: Menggabungkan fitur (PCA)  
* Membuat fitur baru

###**E. Overfitting**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Model terlalu kompleks, belajar noise dalam data, dimana performa bagus di training data, tetapi buruk di data baru.  

Solusi:
* Simplify model (kurangi parameter)
* Regularization (membatasi kompleksitas)
* Dapatkan lebih banyak data
* Kurangi noise

###**F. Underfitting**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Model terlalu sederhana, tidak bisa menangkap pola data, dimana performa buruk baik di training maupun test data.

Solusi:
* Gunakan model lebih kompleks
* Feature engineering
* Kurangi regularization

# **6. Testing dan Validasi**
---
###**A. Train-Test Split**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Memisahkan data menjadi training set (80%) dan test set (20%) dan test set mengestimasi generalization error (out-of-sample error).

###**B. Hyperparameter Tuning dan Model Selection**
* Validation set: Dataset terpisah untuk tuning hyperparameter  
* Cross-validation: Teknik untuk evaluasi model yang lebih robust  
* K-fold cross-validation: Membagi data menjadi K subset, train K model

###**C. Data Mismatch**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ketika training data berbeda dari data produksi  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Solusi:** Train-dev set untuk membedakan overfitting vs data mismatch  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Contoh:** Training dengan gambar web vs gambar mobile app

# **7. No Free Lunch Theorem**
---
* Tidak ada model yang terbaik untuk semua masalah.
* Setiap model membuat asumsi tertentu tentang data.
* Harus mengevaluasi beberapa model untuk menemukan yang terbaik untuk masalah spesifik.

# **9. Teori Penting dalam Chapter 1**
---
###**A. Bias-Variance Tradeoff**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Bias Error:** Kesalahan karena asumsi model terlalu sederhana (underfitting)  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Variance Error:** Kesalahan karena model terlalu sensitif terhadap fluktuasi kecil dalam training data (overfitting)  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Tradeoff**: Model kompleks → variance tinggi, bias rendah; Model sederhana → variance rendah, bias tinggi

###**B. Curse of Dimensionality**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Ketika jumlah fitur sangat besar relatif terhadap jumlah observasi

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Data menjadi sparse di ruang berdimensi tinggi

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Kebutuhan data meningkat secara eksponensial dengan dimensi

###**C. Regularization Theory**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Teknik untuk mencegah overfitting dengan menambahkan constraint pada model  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**L1 regularization (Lasso):** Menghasilkan sparse solution (banyak koefisien = 0)  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**L2 regularization (Ridge):** Menyusutkan koefisien tanpa menjadikannya nol  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Elastic Net:** Kombinasi L1 dan L2  

###**D. Learning Curve Analysis**
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Plot performa training vs validation terhadap jumlah data  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Membantu diagnosis underfitting/overfitting:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Gap besar:** Overfitting  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Performa rendah keduanya:** Underfitting  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Konvergen:** Model cukup

```python
# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Setting style untuk visualisasi
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
# Membuat dataset sederhana untuk demonstrasi regresi
np.random.seed(42)

# Generate data: ukuran rumah (dalam m²) vs harga (dalam $1000)
X = 2.5 * np.random.randn(100) + 50  # Ukuran rumah: mean=50, std=2.5
error = np.random.randn(100) * 10    # Random error
y = 1.8 * X + 20 + error              # Harga = 1.8*ukuran + 20 + error

# Konversi ke DataFrame
house_data = pd.DataFrame({
    'house_size': X,
    'house_price': y
})

# Visualisasi data
plt.figure(figsize=(12, 6))
plt.scatter(X, y, alpha=0.7, s=50, edgecolors='black', linewidth=0.5)
plt.xlabel('Ukuran Rumah (m²)', fontsize=12)
plt.ylabel('Harga Rumah ($1000)', fontsize=12)
plt.title('Hubungan Ukuran Rumah vs Harga', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Menambahkan garis regresi sederhana
z = np.polyfit(X, y, 1)
p = np.poly1d(z)
plt.plot(sorted(X), p(sorted(X)), "r--", linewidth=2, label=f'y = {z[0]:.2f}x + {z[1]:.2f}')
plt.legend(fontsize=12)
plt.show()

print("Dataset contoh:")
print(house_data.head())
print(f"\nStatistik deskriptif:\n{house_data.describe()}")
```
